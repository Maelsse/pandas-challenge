```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_csv = "Resources/schools_complete.csv"
student_data_csv = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_csv)
student_data_df = pd.read_csv(student_data_csv)

# Combine the data into a single dataset.  
merged_data = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
merged_data
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Creating Lists
total_schools = len(school_data_df['school_name'].unique())
total_students = len(student_data_df['student_name'])
total_budget = school_data_df['budget'].sum()
avg_math_score = student_data_df['math_score'].mean()
avg_reading_score = student_data_df['reading_score'].mean()
pass_math = student_data_df.loc[student_data_df['math_score'] >= 70]['math_score'].count()
pass_read = student_data_df.loc[student_data_df['reading_score'] >= 70]['reading_score'].count()
per_pass_math = pass_math / total_students * 100
per_pass_read = pass_read / total_students * 100
overall_pass = student_data_df[(student_data_df['math_score'] >= 70) & (student_data_df['reading_score'] >= 70)]['student_name'].count()/total_students * 100

#Making Sure everythings correct
# print(total_schools)
# print(total_students)
# print(total_budget)
# print(avg_math_score)
# print(avg_reading_score)
# print(pass_math)
# print(pass_read)
# print(per_pass_math)
# print(per_pass_read)
#print(overall_pass)

#Data Frames

district_summary_df = pd.DataFrame({"Total Schools": [total_schools],
                                  "Total Students": [total_students],
                                  "Total Budget": [total_budget], 
                                  "Average Math Score": [avg_math_score], 
                                  "Average Reading Score": [avg_reading_score], 
                                  "Per Passing Math": [per_pass_math], 
                                  "Per Passing Reading": [per_pass_read],
                                  "Overall Passing Rate": [overall_pass]
                                 })
# Some formatting

district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.2f}".format)
district_summary_df["Per Passing Math"] = district_summary_df["Per Passing Math"].map("{:.2f}".format)
district_summary_df["Per Passing Reading"] = district_summary_df["Per Passing Reading"].map("{:.2f}".format)
district_summary_df["Overall Passing Rate"] = district_summary_df["Overall Passing Rate"].map("{:.2f}".format)
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Name
sch_name = merged_data.set_index('school_name').groupby(['school_name'])

#School Types
sch_type = merged_data.set_index('school_name')['type']

#Total Students
sch_students = sch_name["Student ID"].count()

#Total School Budget
sch_budget = school_data_df.set_index('school_name')['budget']

#Per Student Budget
stu_budget = sch_budget / sch_students

#Average Math Score
avg_math = sch_name["math_score"].mean()

#Average Reading Score
avg_read = sch_name["reading_score"].mean()

# % Passing Math
passed_math = merged_data.loc[merged_data["math_score"] >= 70].groupby('school_name')['Student ID'].count()/sch_students

# % Passing Reading
passed_reading = merged_data.loc[merged_data["reading_score"] >= 70].groupby('school_name')['Student ID'].count()/sch_students
                                                                                                    
# Overall Passing Rate
overall_passing = (passed_math + passed_reading)/2
```

```python
# School Summary Data Frame

sch_summary = pd.DataFrame({
    "School Type": [sch_type],
    "Total Students": [sch_students],
    "School Budget": [sch_budget],
    "Student Budget": [stu_budget],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_read],
    "% Passing Math": [passed_math],
    "% Passing Reading": [passed_reading],
    "Overall Passing Rate": [overall_passing],
})

# Rearranging Order

sch_summary = sch_summary[["School Type",
                           "Total Students",
                           "School Budget",
                           "Student Budget",
                           "Average Math Score",
                           "Average Reading Score",
                           "% Passing Math",
                           "% Passing Reading",
                           "Overall Passing Rate"
                           
                          ]]
# Formatting

# sch_summary['Total Students'] = sch_summary['Total Students'].map("{:,}".format)
# sch_summary['School Budget'] = sch_summary['School Budget'].map("{:,}".format)
# sch_summary['Student Budget'] = sch_summary['Student Budget'].map("{:.2f}".format)
# sch_summary['Average Math Score'] = sch_summary['Average Math Score'].map("{:.1f}".format)
# sch_summary['Average Reading Score'] = sch_summary['Average Reading Score'].map("{:.1f}".format)
# sch_summary['% Passing Math'] = sch_summary['% Passing Math'].map("{:.2%}".format)
# sch_summary['% Passing Reading'] = sch_summary['% Passing Reading'].map("{:.2%}".format)
# sch_summary['Overall Passing Rate'] = sch_summar['Overall Passing Rate'].map("{:.2%}".format)

sch_summary
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
math_9 = merged_data.loc[merged_data["grade"] == '9th'].groupby('school_name')['math_score'].mean()
math_10 = merged_data.loc[merged_data["grade"] == '10th'].groupby('school_name')['math_score'].mean()
math_11 = merged_data.loc[merged_data["grade"] == '11th'].groupby('school_name')['math_score'].mean()
math_12 = merged_data.loc[merged_data["grade"] == '12th'].groupby('school_name')['math_score'].mean()

grade_math = pd.DataFrame({"9th Grade": [math_9],
                           "10th Grade": [math_10],
                           "11th Grade": [math_11],
                           "12th Grade": [math_12],
                          })

grade_math
```
